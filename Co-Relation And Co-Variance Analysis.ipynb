```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sbs

df=pd.read_csv('student-mat.csv')
df.head()
df.info()
dfn=df[['traveltime','studytime']]
dfn.head()
x=dfn['traveltime']
y=dfn['studytime']
sbs.lineplot(x=x, y=y, dashes=True)
plt.show()
```
